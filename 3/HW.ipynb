```python
from math import sin
from matplotlib import pyplot as plt
import numpy as np


# the real function y = 2.3 * x^2
C = 2.34 # the real unknown parameter of our function

## generate data with noise
X = np.random.random(300)*3
Y = 3*(0.6-np.random.random(300)) + C*X**2

S1 = sum( y for x,y in zip(X,Y) )
S2 = sum( x**2 for x in X)
print ("C =", S1/S2)

plt.plot(X,Y,'.')
plt.show()



## plot the cost function
F = []
P = []
c = 0.
while c < 4.:
	P.append(c)
	F.append( sum( (c*x**2 - y)**2 for x,y in zip(X,Y)) )
	c += 0.01

plt.plot(P,F,'r-')
plt.grid(True)
plt.show()
```
